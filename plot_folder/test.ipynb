```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

# Example RGB image
rgb_image = np.random.randint(0, 255, size=(200, 200, 3), dtype=np.uint8)

# Create a subplot
fig, ax = plt.subplots()

# Display the RGB image
ax.imshow(rgb_image)


# Define ellipse parameters
center = (rgb_image.shape[1]//2, rgb_image.shape[0]//2)  # Center of the ellipse
width = 0  # Major axis
height = 100  # Minor axis
angle = 90  # Angle of rotation (in degrees)

# Create the ellipse patch
ellipse_patch = Ellipse(center, width, height, angle=angle, fill=False, edgecolor='red',linewidth=2)

# Add the ellipse patch to the subplot
ax.add_patch(ellipse_patch)

# Set plot limits
ax.set_xlim(0, rgb_image.shape[1])
ax.set_ylim(rgb_image.shape[0], 0)

# Show plot
plt.show()
```
