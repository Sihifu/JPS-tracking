```python
from PIL import Image
import matplotlib.pyplot as plt

def merge_images(image1_path, image2_path, output_path, scale_factor=1):
    # Open images
    img1 = Image.open(image1_path)
    img2 = Image.open(image2_path)

    # Scale images
    width, height = img1.size
    new_width = int(width * scale_factor)
    new_height = int(height * scale_factor)
    img1 = img1.resize((new_width, new_height))
    img2 = img2.resize((new_width, new_height))

    # Create a new image with the width doubled to accommodate both images
    merged_img = Image.new('RGB', (new_width * 2, new_height))

    # Paste images side by side
    merged_img.paste(img1, (0, 0))
    merged_img.paste(img2, (new_width, 0))

    # Save the merged image
    merged_img.save(output_path)

    # Show the merged image
    plt.imshow(merged_img)
    plt.axis('off')
    plt.show()
    
# Example usage
image1_path = '/Users/Hoang_1/Desktop/Master_Arbeit/Thesis_VS/Images/cover_1.png'
image2_path = '/Users/Hoang_1/Desktop/Master_Arbeit/Thesis_VS/Images/cover_2.png'
output_path = 'merged_images.png'
merge_images(image1_path, image2_path, output_path, scale_factor=0.5)
```
